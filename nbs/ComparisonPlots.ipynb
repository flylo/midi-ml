```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
```

```python
nb = [0.606, 0.663, 0.6985]
```

```python
l2 = [0.852, 0.8599, 0.873]
```

```python
l1 = [0.804, 0.851, 0.889]
```

```python
fm = [0.615, 0.6552, 0.722]
```

```python
nt = [0.682, 0.693, 0.717]
```

```python
lda = [0.704, 0.7514, 0.7984]
```

```python
model_names = []
metric_names = []
metric_values = []
```

```python
model_names = model_names + ["Multinomial NB" for i in range(3)]
metric_names = metric_names + ["accuracy" for i in range(3)]
metric_values = metric_values + nb
```

```python
model_names = model_names + ["LDA" for i in range(3)]
metric_names = metric_names + ["accuracy" for i in range(3)]
metric_values = metric_values + lda
```

```python
model_names = model_names + ["L1 Logistic Regression" for i in range(3)]
metric_names = metric_names + ["accuracy" for i in range(3)]
metric_values = metric_values + l1
```

```python
model_names = model_names + ["L2 Logistic Regression" for i in range(3)]
metric_names = metric_names + ["accuracy" for i in range(3)]
metric_values = metric_values + l2
```

```python
model_names = model_names + ["Logistic Factorization Machine" for i in range(3)]
metric_names = metric_names + ["accuracy" for i in range(3)]
metric_values = metric_values + fm
```

```python
model_names = model_names + ["n-Tuple Classifier" for i in range(3)]
metric_names = metric_names + ["accuracy" for i in range(3)]
metric_values = metric_values + nt
```

```python
metric_df = pd.DataFrame({"model": model_names,
"metric": metric_names,
"values": metric_values})
accuracy_df = metric_df[metric_df.metric == "accuracy"]
accuracy_df.columns = ["metric", "Model", "Accuracy Score"]
```

```python
plt.figure(figsize=(8,6))
sns.set_style("whitegrid")
g = sns.stripplot(x="Model",
              y="Accuracy Score",
              data=accuracy_df,
              size=8,
              color="gray")
for item in g.get_xticklabels():
    item.set_rotation(90)

plt.title("Accuracy Score - All Models")
plt.show()
```
